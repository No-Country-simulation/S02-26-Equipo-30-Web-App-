```python
import pandas as pd

# 1锔 Cargar dataset
df = pd.read_csv("../data/horsetrust_database_index.csv")

# 2锔 Establecer horse_id como 铆ndice
df.set_index("horse_id", inplace=True)

print("Archivo cargado correctamente ")

# 3锔 Funci贸n para consultar trust score + columnas relevantes + interpretaci贸n
def get_horse_info(horse_id: str):
    horse_id = horse_id.strip()
    
    # Verificamos si existe
    if horse_id not in df.index:
        return f"Horse ID '{horse_id}' no encontrado."
    
    # Columnas que queremos mostrar
    columns_to_show = [
        "height_m",
        "weight_kg",
        "length_m",
        "max_speed_kmh",
        "h_career_races",
        "h_days_since_last_race",
        "l_asking_price_usd",
        "l_price_vs_market_ratio",
        "s_disputes",
        "s_num_listings",
        "vet_total_exams",
        "vet_major_issues"
    ]
    
    horse_data = df.loc[horse_id, columns_to_show].copy()
    
    # Obtener trust score y multiplicar por 100
    trust_score = df.loc[horse_id, "horse_trust_score"] * 100
    horse_data["horse_trust_score_%"] = round(trust_score, 2)
    
    # Interpretaci贸n
    if trust_score < 70:
        interpretation = "Se recomienda revisar m谩s datos del caballo. ndice bajo"
    elif 70 <= trust_score < 87:
        interpretation = "Confiable"
    else:
        interpretation = "Excelente"
    
    horse_data["Interpretaci贸n"] = interpretation
    
    return horse_data

# 4锔 Uso
horse_id_input = input("Ingrese el horse_id: ")
horse_info = get_horse_info(horse_id_input)

print("\n Informaci贸n del caballo:")
print(horse_info)
```
